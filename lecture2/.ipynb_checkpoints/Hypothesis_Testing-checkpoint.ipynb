First, we sample fake datasets from normal distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
_m1, _s1 = 10, 18
d1 = np.random.normal(_m1, _s1, 1000)

_m2, _s2 = 13, 18
d2 = np.random.normal(_m2, _s2, 1000)
```

### Compute means and standard deviations

For now, we can use the useful, in-built numpy functions.

```python
print(d1.mean(), d1.std())
print(d2.mean(), d2.std())
```

## Visualize datasets

Line plots: Good in many cases (e.g. timeseries applications as we saw in Lecture 1), but not great here...

```python
plt.plot(d1, label="d1")
plt.plot(d2, label="d2")
plt.legend()
plt.show()
```

Histograms: Shows the frequency of data when the x-axis is 'binned' or bucketed into several segments. Use `density=True` if you want to change the frequency into a probability distribution (i.e. normalize with respect to the number of data points)

```python
plt.hist(d1, bins=30, label="d1")
plt.hist(d2, bins=30, label="d2")
plt.legend()
plt.show()
```

Fit a normal/bell/Gaussian curve over it to visualize the 'expected' dataset. An expectation of a random variable is essentially the weighted distribution of what value you expect to get when you have sampled the whole population:

$$E[X] = \sum_{i=0}^\infty x_ip(x_i)$$
$$E[X] = \int_{i=0}^\infty x_ip(x_i) dx$$

We can also similarly get the expected variance or standard distribution of this variable $X$

We can approximate it wth the sample mean and sample standard deviation.

$$\mu=\frac{1}{N}\sum_{i=0}^N x_i$$
$$\sigma=\sqrt{\frac{\sum_{i=0}^N (x_i-\mu)^2}{N}}$$


If we know that we expect a normal distribution, which is the case here, we can fit it using the computed mean and standard deviation by putting it into the function:

$$\mathcal{N}(x|\mu,\,\sigma^{2})=\frac{1}{\sigma\sqrt{2\pi}}\exp\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right)$$

```python
m = d1.mean()
s = d1.std()

x = np.linspace(d1.min(), d1.max(), 100)
y = stats.norm.pdf(x,m,s)

plt.hist(d1, bins=30, density=True)
plt.plot(x,y, color="red")
plt.show()
```

Do this for both datasets

```python
x = np.linspace(d1.min(), d1.max(), 100)

m1, s1 = d1.mean(), d1.std()
m2, s2 = d2.mean(), d2.std()

y1 = stats.norm.pdf(x,m1,s1)
y2 = stats.norm.pdf(x,m2,s2)

plt.hist(d1, bins=30, density=True, alpha=0.7)
plt.hist(d2, bins=30, density=True, alpha=0.7)

plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```

## T-Tests

A t-test is used to determine whether the difference of means of two distributions is zero. You can also do a T-test with a single distribution, and then is determines whether or not the mean of the distribution is zero.

H0: Null hypothesis: "The distribution means of the two datasets are the same."

If the p-value is less than some threshold, usually 0.05, we can reject the null hypothesis. Else, we cannot reject it (note that this is _not_ the same as 'accepting' it).

Note that, in general, we have these four assumptions:
1. The data is continuous. 
2. The sample data is randomly sampled from some underlying population.
3. Homogeneity of variance (the variance in both datasets are similar).
4. The sample distribution is approximately normal.

There are other t-tests/variants that can account for these assumptions:
- One-tailed/Two-tailed
- Paired/unpaired
- Parametric/non-parametric
- ANOVA (>2 datasets)
- ...

(Perfect time to Google! Example: https://www.investopedia.com/terms/t/t-test.asp)

First, we can plot our _expected_ datasets:

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.fill_between(x, np.minimum(y2,y1), color="green")
```

```python
stats.ttest_ind(d1, d2)
```

As a rule of thumb for significance, we can plot the bar plot of this with the error bars equal to the standard error:
$$SE = \frac{\sigma}{\sqrt N}$$

If the error bars overlap, they are generally indicitive of being not significant.

```python
plt.bar(0, m1)
plt.bar(1, m2)
se1 = s1/np.sqrt(len(d1))
se2 = s2/np.sqrt(len(d2))
plt.errorbar(0, m1, se1, capsize=10, color="black")
plt.errorbar(1, m2, se2, capsize=10, color="black")
```
