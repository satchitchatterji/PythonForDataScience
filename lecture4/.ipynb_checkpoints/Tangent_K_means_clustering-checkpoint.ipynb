## Generate clusters of data
First, let's play around with a simulated dataset of blobs of data. The data is generated with respect to two variables/features _x1_ and _x2_. For example, x1 can be weight and x2 can be furriness. In general, data can have any number of _n_ features. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Blobs: In this case, defined as 2D Gaussian clusters
N1, N2, N3 = 100, 85, 106

# isotropic gaussians for simplicity
mu1, mu2, mu3 = np.array([1.0,1.0]), np.array([3.0,3.0]), np.array([4.0,-1.0])
s = 0.3
S1 = np.eye(2)*s
S2 = np.eye(2)*s
S3 = np.eye(2)*s

# generate class-specific data points
X1 = np.random.multivariate_normal(mu1, S1, size=N1)
X2 = np.random.multivariate_normal(mu2, S2, size=N2)
X3 = np.random.multivariate_normal(mu3, S3, size=N3)

# plot the,
plt.scatter(X1.T[0], X1.T[1], label="Class 1")
plt.scatter(X2.T[0], X2.T[1], label="Class 2")
plt.scatter(X3.T[0], X3.T[1], label="Class 3")

plt.legend()

plt.title("Generated data")
# x1 and x2 are features,
# e.g. weight and furriness
plt.xlabel("x1")
plt.ylabel("x2")

plt.show()
```

If x1 is furriness and x2 is weight, we can think of the first class, for example, as golden retrievers, the second as pugs and the third as persian cats (convince yourself of this analogy). However, in real datasets, we sometimes do not have these 'labels' -- we just have the data. So let's concatenate all three classes into a single dataset with no labelled distinction between classes.

```python
X = np.concatenate([X1, X2, X3])
print(X.shape)
```

We can plot a scatter plot to show that all classes now should look the same (apart from its x1 and x2 positions). As humans, we can still see 3 clusters, but we need an algorithm to label these points into those 3 clusters. In other cases there may be even more clusters that can be hard to distinguish, and in higher dimensions, we may not even be able to visualize clusters! 

```python
plt.scatter(X[:,0], X[:,1])
plt.title("Generated data (Unlabelled)")
plt.xlabel("x1")
plt.ylabel("x2")
```

## K-means Clustering

- The most popular clustering method
- Iterative in nature
- Algorithm:
    - Initialize random "centroids" (a "centroid" is the numerical center of the cluster)
    - Do until convergence:
        - "_Expectation step_": Assign data points to groups based on the closest centroid.
        - "_Maximization step_": Re-compute centroids to be the mean of the data points in each direction.
    - _Convergence_: The most common sign is to wait until the centroids 'settle' -- i.e. wait until they do not change from iteration to iteration.

A naive implementation is below. Try and convince yourself that this function does indeed implement the algorithm above.

```python
def distance(a, b):
    """ Euclidean distance between vectors a and b """
    return np.linalg.norm(a - b)

def k_means(data, k, max_iter=10, plot_figs=True):
    """ Performs k-means clustering on data
    NOTE: This implementation has stability issues

    Arguments:
        data: n-dimensional data of shape (n_points, n_features)
        k: number of clusters to find
        max_iter: maximum number of iterations in case of non-convergence
        plot_figs: whether or not to plot figures at each iteration
    Returns:
        classes: dictionary of {centroid_position : list_of_cluster_members}
        mean_distance: mean centroid distance to points
    """
    # initialize centroids and create object to store data
    centroids = np.random.uniform(data.min(), data.max(), (k, data.shape[1]))
    classes = {tuple(c):[] for c in centroids}
    mean_distance = None

    # iterate until convergence
    for iter in range(max_iter):
        # clear out class members (needs to be recalculated)
        classes = {c:[] for c in classes.keys()}

        # Expectation step:
        # for each of the points in the dataset
        #   calculate the closest centroid C
        # append the point to the cluster whose centroid is C
        for point in X:
            closest_centroid = None
            closest_distance = None
            for centroid in classes.keys():
                if closest_distance is None or distance(centroid, point) < closest_distance:
                    closest_centroid = centroid
                    closest_distance = distance(centroid, point)
            
            classes[closest_centroid].append(point)

        # Store previous centroids to check convergence later
        prev_centroids = set(classes.keys())        

        # Maximization step:
        # update the centroids to be the mean of the points in its cluster
        for centroid, points in classes.items():
            if points:
                del classes[centroid]
                centroid = tuple(np.mean(points, axis=0))
                classes[centroid] = points

        # plot the predicted clusters for this iteration
        if plot_figs:
            for i, (centroid, points) in enumerate(classes.items()):
                points = np.array(points)
                if len(points) != 0:
                    plt.scatter(points[:,0], points[:,1], label="class "+str(i))
                plt.scatter(centroid[0], centroid[1], color="black", marker="x")
            plt.title("K-means iteration "+str(iter))
            plt.legend()
            plt.show()

        # calculate mean centroid distance
        distances = []
        for centroid, points in classes.items():
            for point in points:
                distances.append(distance(centroid, point))
        mean_distance = np.mean(distances)

        # Check convergence:
        # if the clusters are at the same position as before
        # the algorithm has converged
        if prev_centroids == set(classes.keys()):
            print(f"Converged! After {iter+1} iterations")
            return classes, mean_distance

    # if max_iterations have been done
    print(f"Could not converge after {max_iter} iterations!")
    return classes, mean_distance

classes, mean_distance = k_means(X, k=10, max_iter=40, plot_figs=True)
```

## But how many groups? The 'elbow' method
In general, real-world datasets, we generally do not have access to the true number of clusters. These clusters can be subjective too. Thus, one way to analyse clustering is called the 'elbow' method: plot the mean point-to-centroid distance over all clusters. We then pick the number of clusters that occur near the 'elbow' of the resulting curve. This doesn't always give a clear result, but is a good first attempt.

> Q) What happens as the number of clusters increase?

```python
distances = []
for i in range(1,11):
    print(f"k={i}...")
    classes, mean_distance = k_means(X, k=i, max_iter=40, plot_figs=False)
    distances.append(mean_distance)

plt.figure(figsize=(7.5,5))
plt.ylabel("Mean cluster distance")
plt.xlabel("# clusters")

plt.plot(list(range(1,11)), distances)
plt.show()
```

### Drawbacks and things to keep in mind
- We use Euclidean distance as a distance measure
    - This is prone to effect from outliers
    - Generally only good for isotropic (no covariance), similar-sized clusters. 
    - Sometimes k-means is slow to converge, depending on the dataset and initial conditions.

```python
N1, N2, N3 = 500, 500, 1000

mu1, mu2, mu3 = np.array([4.5,4.0]), np.array([1.5,4.0]), np.array([3.0,2.0])

# isotropic gaussians for simplicity
S1 = np.eye(2)*0.1
S2 = np.eye(2)*0.1
S3 = np.eye(2)*0.5

X1 = np.random.multivariate_normal(mu1, S1, size=N1)
X2 = np.random.multivariate_normal(mu2, S2, size=N2)
X3 = np.random.multivariate_normal(mu3, S3, size=N3)

# outlying point
# X3[0] = np.array([10,10]) 

plt.figure(figsize=(7,7))

plt.scatter(X1.T[0], X1.T[1], label="Class 1")
plt.scatter(X2.T[0], X2.T[1], label="Class 2")
plt.scatter(X3.T[0], X3.T[1], label="Class 3")

plt.legend()
plt.title("Generated data")
plt.xlabel("x1")
plt.ylabel("x2")


plt.show()
```

```python
X = np.concatenate([X1, X2, X3])

plt.figure(figsize=(7,7))
plt.scatter(X[:,0], X[:,1])
```

```python
classes, mean_distance = k_means(X, k=3, max_iter=40, plot_figs=True)
```
