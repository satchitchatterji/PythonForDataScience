## Tangent: Overfitting and Underfitting
> Today's question: **How do make sure my model reflects the real world?**

The M=0 model _underfits_: it does not do well either on the underlying function nor the actual data. It does not learn the underlying function and the error on the data is high.

The M=9 model _overfits_: it does very well on the actual data (there is zero error between the line and the points) but it doesn't fit the true function very well. It learns the data too well -- including the noise (which is not ideal), and thus does not generalise well to new, unseen data. 

The M=3 model seems to neither overfit nor underfit, and we can see this by compararing it to the underlying function -- however, it is not always clear what the underlying function actually is.

Let's generate some data!

```python
import numpy as np
import matplotlib.pyplot as plt

true_xs = np.linspace(0, 2*np.pi, 1000)
true_ys = np.sin(true_xs)

xs = np.linspace(0, 2*np.pi, 10)
ys = np.sin(xs)
plt.plot(xs, ys, label="Sampled function")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.show()
```

Generally, real world data has associated noise.

> Q) What kind of (noise) distributions can we expect from real-world data?

```python
noise = np.random.normal(0, .3, len(ys))
ys = np.sin(xs) + noise
plt.plot(xs, ys, label="Sampled function")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.legend()
plt.show()
```

We can now fit polynomials over our attained dataset (the blue curve above, the orange one below).

```python
params = np.polyfit(xs, ys, deg=1)
line_fit = np.polyval(params,xs)
plt.plot(xs, line_fit, label="Fit function")
plt.plot(xs, ys, label="Sampled function")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.legend()
plt.show()
```

#### Define evaluation measure: Mean Squared Error
How good is our fit with respect to our data?

> Q) Why use MSE here?

> Q) Others?

$$MSE=\frac{1}{N}\sum^{N}_{n=1}(a-b)^2$$

Where $a$ are the predictions and $b$ are the values from data (or vice versa).

```python
def mse(a, b):
    return ((a - b)**2).mean()

mse(line_fit, ys)
```

What about higher-order polynomials?

```python
params = np.polyfit(xs, ys, deg=2)
line_fit = np.polyval(params,xs)
plt.plot(xs, line_fit, label="Fit function")
plt.plot(xs, ys, label="Sampled function")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.legend()
plt.show()
```

```python
mse(line_fit, ys)
```

```python
params = np.polyfit(xs, ys, deg=3)
line_fit = np.polyval(params,xs)
plt.plot(xs, line_fit, label="Fit function")
plt.plot(xs, ys, label="Sampled function")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.legend()
plt.show()
```

Let's compute MSE distances over different model parameters and plot it.

```python
distances = []
for i in range(1, 15):
    params = np.polyfit(xs, ys, deg=i)
    line_fit = np.polyval(params, xs)
    distances.append(mse(line_fit, ys))

plt.xlabel("degree")
plt.ylabel("MSE")
plt.plot(list(range(1,15)), distances)
plt.show()
```

We see that the MSE decreases with the 'flexibility' of the model. But why is a smaller error not necessarily better?

>Q) Think of this with respect to the M=9 model we saw at the very beginning.

## Approaches to fixing this
### Validation sets!

One solution (and the most common) to aleviate this issue of what model to choose is to split the training dataset into two sets -- one you train with, one you 'validate' with. If you have a lot of points, you can split it into two sets of e.g. 80% train and 20% validation.

In this case, since there are so few data points and I'm generating the datapoints anyway, I'll assume there exists a second set of real-world points given to me. They will be from the same distribution and the training set we had earlier.

```python
xs_val = np.linspace(0, 2*np.pi, 10)
ys_val = np.sin(xs_val)
noise = np.random.normal(0, .3, len(ys_val))
ys_val += noise

plt.plot(xs, ys, label="Sampled function (test)")
plt.plot(xs_val, ys_val, label="Sampled function (validation)")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.legend()
plt.show()
```

We now compute validation and training set curves -- the MSE error between the line fit and the training set, and the MSE error between the line fit and the validation set. The validation set acts as a representation of unseen data points, for example, those when you actually use it in an application.

```python
train_distances = []
val_distances = []

for i in range(1, 15):
    params = np.polyfit(xs, ys, deg=i)
    line_fit = np.polyval(params, xs)
    train_distances.append(mse(line_fit, ys))
    val_distances.append(mse(line_fit, ys_val))

plt.xlabel("degree")
plt.ylabel("MSE")
plt.plot(list(range(1,15)), train_distances, label="Train curves")
plt.plot(list(range(1,15)), val_distances, label="Validation curves")
plt.legend()
plt.show()
```

Analysis

Far left: _Underfitting_: Does not perform well on the training or validation sets.

Far right: _Underfitting_: Performs well on the training set, but does not generalise well to new data points. 

The best model (subjectively) is where the validation set is lowest. In this case, we see that the model with degree 3 is indeed the best one, and this fits our intuition perfectly. This problem gets harder with more complex models, and the need for plotting these curves gets more and more important.

Let's fit an M=3 model and see how well it fits. What can we do better?


```python
params = np.polyfit(xs, ys, deg=3)
line_fit = np.polyval(params,xs)
plt.plot(xs, line_fit, label="Fit function")
# plt.plot(xs, ys, label="Sampled function")
plt.plot(true_xs, true_ys, label="True function", color="r", linestyle="--")
plt.legend()
plt.show()
```
